### Usando dbpedia

````sparql
%endpoint http://dbpedia.org/sparql
%display table
%show all
````

#### Principais institutos/organizações relacionados ao tema de câncer de mama

````sparql
select ?search
where{
    ?search dct:subject dbc:Breast_cancer 
}
````

#### Pessoas que sobreviveram à doênça

````sparql
select ?search
where{
    ?search dct:subject dbc:Breast_cancer_survivors . 
}
````

#### Cantoras Alemãs que sobreviveram ao câncer de mama

````sparql
select ?search
where{
    ?search dct:subject dbc:Breast_cancer_survivors .#search for survivors
    ?search dct:subject dbc:German_female_singers    #among the survivors, who was a female German Singer
}
````

#### Pessoas que morreram de câncer de mama

````sparql
select ?person
where{
    ?person dbo:deathCause dbr:Breast_cancer 
}

````

#### Toṕico relacionado ao câncer de mama que foi causa de morte de alguma pessoa
Esta consulta retorna dois resultados, o primeiro é o tópico (Male Breast Cancer), que foi responsável pela morte de uma pessoa. Esta por sua vez era um assassino serial.

````sparql
select ?search, ?person
where{
    ?search  dct:subject dbc:Breast_cancer . #subjects related to breast_cancer
    ?person dbp:cause ?search                #person whose death cause is a subject related to breast cancer
  
}
````
